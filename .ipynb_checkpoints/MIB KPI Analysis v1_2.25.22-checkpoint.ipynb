## Process SAT files

```python
'''Import data files'''
path = 'data/sats'
csv_files = glob.glob(path + '/*.csv')
var_dict = {}

for file in csv_files:
    key = file
    value = pd.read_csv(file)
    var_dict[key] = value
```

```python
'''Drop columns that are not used'''
drop_columns = ['GPA', 'AP/IB/Honors', 'Program Strength', 'Highest PreACT', 'Highest Aspire', 'Highest PLAN', 'Decile']
for key, item in var_dict.items():
    for column in item.columns:
        if column in drop_columns:
            item.drop([column], axis = 1, inplace = True)
    msno.matrix(item)
```

```python
'''Convert Student IDs to int'''
for key, item in var_dict.items():
    for column in item.columns:
        if column == 'Student ID':
            item[column].replace(np.nan, '00000000', inplace = True)
            item[column] = item[column].astype('int64')
            print(item.info())
```

```python
'''Combine and standardize various iterations of SAT scores'''
for key, item in var_dict.items():
    item['Highest SAT EBRW'] = np.where(np.isnan(item['Highest SAT EBRW']), 
                                    (item['Highest SAT CR (taken before Mar 2016)'] + 
                                     item['Highest SAT WR (taken before Mar 2016)'])/2,
                                    item['Highest SAT EBRW'])
    item['Highest SAT M'] = np.where(np.isnan(item['Highest SAT M']), 
                                    item['Highest SAT M (taken before Mar 2016)'],
                                    item['Highest SAT M'])
    item['Highest PSAT'] = np.where(np.isnan(item['Highest PSAT']), 
                                    item['Highest PSAT (taken before Oct 2015)'],
                                    item['Highest PSAT'])
    item.drop(['Highest SAT CR (taken before Mar 2016)', 'Highest SAT WR (taken before Mar 2016)',
               'Highest SAT M (taken before Mar 2016)', 'Highest PSAT (taken before Oct 2015)', 
              'Highest Comb. SAT 2400'], axis = 1, inplace = True)
```

```python
for key, item in var_dict.items():
    msno.matrix(item)
```

```python
'''Combine all untransformed data into single dataframe'''
keys_list = list(var_dict)

for key, item in var_dict.items():
        df1 = item
        next_index = (keys_list.index(key) + 1)
        if next_index < len(keys_list):
            next_index_item = keys_list[next_index]
            df2 = var_dict[next_index_item]
            if keys_list.index(key) == 0:
                test_df_concat = pd.concat([df1,df2], axis=0, ignore_index=True)
            else: test_df_concat = pd.concat([test_df_concat, df2], axis=0, ignore_index=True)

test_df_concat['Student ID'] = test_df_concat['Student ID'].astype('string')
test_df_concat
```

## Process grade files

```python
'''Import grades file'''
mk_grades = pd.read_csv('data/grades/mk_grades.csv')
mh_grades1 = pd.read_csv('data/grades/mh_grades1.csv')
mh_grades2 = pd.read_csv('data/grades/mh_grades2.csv')
mh_grades3 = pd.read_csv('data/grades/mh_grades3.csv')
```

```python
'''Rename columns'''
grades.rename({'Student > LASID': 'lasid', 'Student > YOG': 'yog', 'SchoolYear > Year' : 'year', 'Grade': 'grade_level',
                  'SchoolCourse > CrsNo' : 'course_no', 'SchoolCourse > Description': 'course_desc', 'Staff > Name' : 'staff',
                  'FinalGrade' : 'final_grade', 'MasterSchedule > Term' : 'term', 'MP1' : 'mp1', 'MP2' : 'mp2', 'MP3' : 'mp3',
                  'MP4' : 'mp4', 'MP1 Exam' : 'mp1_exam', 'MP2 Exam_Project' : 'mp2_exam', 'MP3 Exam' : 'mp3_exam', 
                  'MP4 Exam_Project' : 'mp4_exam', 'Exam_Average' : 'exam_average'}, axis = 1, inplace = True)
grades
```

```python
'''Replace text values with numeric data in numeric columns'''
for column in grades.columns:
    grades[column].replace(['WP', 'DP', 'P', 'PS', 'p'], value = 70, inplace = True)
    grades[column].replace(['WF', 'DF', 'F'], value = 60, inplace = True)
    grades[column].replace(['X', 'W', 'I', 'AU', 'A', 'x'], value = np.nan, inplace = True)
    if column == 'grade_level':
        grades[column].replace(['8', 8,'PG'], value = 20, inplace = True)
```

```python
'''Convert numeric columns to numeric dtype'''
float_columns = ['final_grade', 'mp1', 'mp2', 'mp3', 'mp4', 'mp1_exam', 'mp2_exam', 'mp3_exam',\
                 'mp4_exam','exam_average']
int_columns = ['yog', 'year', 'grade_level']

for column in float_columns:
    grades[column] = grades[column].astype('float64')
for column in int_columns:
    grades[column] = grades[column].astype('int64')
```

```python
grades.info()
```

```python
'''Create seperate df for each deparment'''
ss_df = grades[grades['course_no'].str.contains(r'^SS\d')]
sc_df = grades[grades['course_no'].str.contains('SC')]
mt_df = grades[grades['course_no'].str.contains('MT')]
en_df = grades[grades['course_no'].str.contains('EN')]
ss_df
```

```python
'''Create seperate dataframes for each department and grade level'''
departments = ['ss', 'sc', 'mt', 'en']
grade_levels = ['9','10','11','12']
data_frames = [ss_df, sc_df, mt_df, en_df]
df_dict = {}

for department in departments:
    for grade_level in grade_levels:
        key = department + grade_level + '_df'
        df = department + '_df'
        value = locals()[df].loc[locals()[df]['grade_level'] == int(grade_level)]
        df_dict[key] = value
        df_dict[key].columns = df_dict[key].columns+ '_' + grade_level + department
```

```python
df_dict
```

```python
'''Change lasid column names for merge'''
for key, item in df_dict.items():
    item.columns = ['lasid' if 'lasid' in x else x for x in item.columns]
```

```python
'''Find unique courses'''
for key, item in df_dict.items():
    for column in item.columns:
        if 'course_desc' in column:
            print(item[column].unique())
```

```python
'''Keep main courses in each department'''
keep_column = ['US History', 'World History', 'IB History', 'Biology', 'Chemistry', 'Environmental', 'Physics',
              'Algebra', 'Geometry', 'Calculus', 'Precalculus', 'English', 'Lit', 'Lang']

for key, item in df_dict.items():
    for column in item.columns:
        if 'course_desc' in column:
            mask = item[column].apply(lambda x: any(course for course in keep_column if course in x))
            item = item[mask]
            df_dict[key] = item
```

```python
'''Keep only main courses in each department per studentID'''
for key, item in df_dict.items():
    item.drop_duplicates(subset='lasid', keep='first', inplace=True)
    for column  in item.columns:
        if column == 'lasid':
            print(len(item), len(item[column].unique()))
```

```python
'''Merge dfs back into single frame'''

df = df_dict[list(df_dict.keys())[0]]

for key, item in df_dict.items():
    if key != 'ss9_df':
        df_merged = pd.merge(df, item, on=['lasid'], how='outer')
        df = df_merged
```

```python
df
```

```python
'''Verify appropriate courses'''
for column in df.columns:
    if 'course_desc' in column:
        print(df[column].unique())
```

```python
df['lasid'] = df['lasid'].astype('string')
```

```python
'''Exclude ODP and AHS students'''
df = df[~df['lasid'].str.contains('ODP')]
df = df[~df['lasid'].str.contains('A')]
df
```

```python
'''Create final DataFrame for grades'''
df.drop_duplicates(keep='first', inplace = True)
df.dropna(axis = 0, thresh = 1, inplace = True)
grades_final_df = df
grades_final_df
```

```python
for column in ap_scores.columns:
    ap_scores[column] = ap_scores[column].astype('string')
```

```python
'''Convert AP Awards to text'''
award_columns = [x for x in ap_scores.columns if 'Award Type' in x]
map_dict = {
        '1.0': 'AP Scholar',
        '2.0': 'AP Scholar with Honor',
        '3.0': 'AP Scholar with Distinction',
        '4.0': 'State AP Scholar',
        '5.0': 'National AP Scholar',
        '6.0': 'National AP Scholar (Canada)',
        '7.0': 'AP International Diploma',
        '8.0': 'DoDEA AP Scholar',
        '9.0': 'International AP Scholar',
        '12.0': 'National AP Scholar (Bermuda)',
        '13.0': 'AP Capstone Diploma',
        '14.0': 'AP Seminar and Research Certificate'
        }

for column in award_columns:
    ap_scores[column] = ap_scores[column].map(map_dict).fillna(ap_scores[column])
```

```python
'''Add .0 to the end of Admin Year 01 entries'''
ap_scores['Admin Year 01'] = ap_scores['Admin Year 01'] + '.0'
ap_scores['Admin Year 01'].unique()
```

```python
ap_scores
```

```python
'''Keep only a single row per student'''
exam_years = ['Admin Year 01',
              'Exam Code 01', 'Exam Grade 01', 'Admin Year 02', 'Exam Code 02', 'Exam Grade 02', 'Admin Year 03',
              'Exam Code 03', 'Exam Grade 03', 'Admin Year 04', 'Exam Code 04', 'Exam Grade 04', 'Admin Year 05', 
              'Exam Code 05', 'Exam Grade 05', 'Admin Year 06', 'Exam Code 06', 'Exam Grade 06', 'Admin Year 07', 
              'Exam Code 07', 'Exam Grade 07', 'Admin Year 08', 'Exam Code 08', 'Exam Grade 08', 'Admin Year 09', 
              'Exam Code 09', 'Exam Grade 09', 'Admin Year 10', 'Exam Code 10', 'Exam Grade 10', 'Admin Year 11', 
              'Exam Code 11', 'Exam Grade 11', 'Admin Year 12', 'Exam Code 12', 'Exam Grade 12', 'Admin Year 13', 
              'Exam Code 13', 'Exam Grade 13', 'Admin Year 14', 'Exam Code 14', 'Exam Grade 14', 'Admin Year 15', 
              'Exam Code 15', 'Exam Grade 15', 'Admin Year 16', 'Exam Code 16', 'Exam Grade 16', 'Admin Year 17', 
              'Exam Code 17','Exam Grade 17']

multi_names = ap_scores['Full Name'].value_counts() > 1
multi_names = multi_names[multi_names].index
counter = 0
ap_temp = pd.DataFrame()

for name in ap_scores['Full Name'].unique():
    if name in multi_names:
        df_temp = ap_scores.loc[ap_scores['Full Name'] == name]
        df_temp['names_max'] = df_temp[exam_years].apply(lambda x: x.isnull().sum(), axis = 1)
        df_temp = df_temp.sort_values(by = ['names_max'], ascending = True)
        df_temp = pd.DataFrame(df_temp.iloc[0].T)
        ap_temp = pd.concat([ap_temp, df_temp], axis = 1, ignore_index = True)
ap_temp
```

```python
'''Create dataframe with single entry per student'''
ap_multiple = ap_temp.T
ap_multiple
```

```python
'''Create dataframe with students that have only one entry'''
single_names = ap_scores['Full Name'].value_counts() == 1
single_names = single_names[single_names].index
ap_single = ap_scores.loc[ap_scores['Full Name'].isin(single_names)]
ap_single.reset_index(inplace = True, drop = True)
ap_single
```

```python
'''Combine dataframes into single frame'''
ap_scores_df = pd.concat([ap_multiple, ap_single], axis = 0)
ap_scores_df.reset_index(inplace = True, drop = True)
ap_scores_df
```

```python
'''Create a column that indicates success in magnet program'''
exam_scores = [x for x in ap_scores_df.columns if 'Exam Grade' in x]
year_list = [x for x in ap_scores_df.columns if 'Admin Year' in x]
award_list = [x for x in ap_scores_df.columns if 'Award Type' in x]
above_three_count = []


ap_scores_df[year_list] = ap_scores_df[year_list].fillna(0).astype('float64')
ap_scores_df[exam_scores] = ap_scores_df[exam_scores].fillna(0).astype('float64')

for index, row in ap_scores_df.iterrows():
    row.fillna(0, inplace=True, axis = 0)
    if (max(row[year_list] > 16) and ('AP Capstone Diploma' in list(row[award_list]) or 'AP Seminar and Research Certificate' in list(row[award_list]))):
        row['success'] = 1
        ap_scores_df.at[index, 'success'] = row['success']
    elif max(row[year_list]) < 17:
        score_list = np.array([row[x] for x in exam_scores])
        total_tests = np.count_nonzero(score_list)
        number_above_three = np.count_nonzero((score_list == 3) | (score_list > 3))
        above_three_count.append(number_above_three)
        above_three_count_avg = sum(above_three_count)/len(above_three_count)
        if number_above_three >= 4 and above_three_count_avg >= 4:
            row['success'] = 1
            ap_scores_df.at[index, 'success'] = row['success']
ap_scores_df['success'].replace(np.nan, 0, inplace = True)
ap_scores_df
```

```python
'''Create dataframe for only students who meet magnet success criteria'''
magnet_success = ap_scores_df.loc[ap_scores_df['success'] == 1.0]
magnet_success.reset_index(inplace = True, drop = True)
magnet_success
```

```python
'''Calculate average number of AP tests passed with 3 or higher for AP Cert students'''
exam_scores = [x for x in magnet_success.columns if 'Exam Grade' in x]
award_list = [x for x in magnet_success.columns if 'Award Type' in x]
magnet_success[exam_scores] = magnet_success[exam_scores].astype('float64')
magnet_success[award_list] = magnet_success[award_list].astype('string')
magnet_success[award_list].replace(pd.NA, 'none', inplace = True)
above_three_count = []

for index, row in magnet_success.iterrows():
    if 'AP Seminar and Research Certificate' in list(row[award_list]):
        score_list = np.array([row[x] for x in exam_scores])
        total_tests = np.count_nonzero(score_list)
        number_above_three = np.count_nonzero((score_list == 3) | (score_list > 3))
        above_three_count.append(number_above_three)
        above_three_count_avg = sum(above_three_count)/len(above_three_count)
print(f'On average, students who earned the AP Seminar and Research Certificate passed {above_three_count_avg } AP exams with a 3 or higher.')
```
